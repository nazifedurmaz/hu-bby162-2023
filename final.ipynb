```python
from google.colab import drive
drive.mount('/content/drive')

dosyaYolu = "/content/drive/MyDrive/Colab Notebooks/veri.txt"

def menu_goster():
    print("Katalog Uygulaması\n")
    print("1. Katalogdaki Eserleri Listele")
    print("2. Yeni Eser Ekle")
    print("3. Eser Sil")
    print("4. Eser Güncelle")
    print("5. Katalogu Temizle")
    print("6. Eser Ara")
    print("0. Çıkış")
    print()

def eserleri_listele():
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
        if not books:
            print("Katalog boş.")
        else:
            for book in books:
                print(book.strip())

def eser_ekle():
    eseradi = input("Eser Adı: ")
    yazari = input("Yazar Adı: ")
    basimyili = input("Basım Yılı: ")
    kitap = f"{eseradi},{yazari},{basimyili}\n"
    with open(dosyaYolu, "a") as file:
        file.write(kitap)
    print("Eser başarıyla eklendi.")

def eser_sil():
    eseradi = input("Silinecek Eserin Adı: ")
    kitaplar = []
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
    with open(dosyaYolu, "w") as file:
        deleted = False
        for kitap in kitaplar:
            if kitap.split(",")[0] != eseradi:
                file.write(kitap)
            else:
                deleted = True
        if deleted:
            print("Eser başarıyla silindi.")
        else:
            print("Eser bulunamadı.")

def eser_guncelle():
    eseradi = input("Güncellenecek Eserin Adı: ")
    kitaplar = []
    with open(dosyaYolu, "r") as file:
        kitaplar = file.readlines()
    with open(dosyaYolu, "w") as file:
        updated = False
        for kitap in kitaplar:
            if kitap.split(",")[0] != eseradi:
                file.write(kitap)
            else:
                yenieseradi = input("Yeni Eser Adı: ")
                yazaradi = input("Yazar Adı: ")
                basimyili = input("Basım Yılı: ")
                guncellenmiskitap = f"{yenieseradi},{yazaradi},{basimyili}\n"
                file.write(guncellenmiskitap)
                updated = True
        if updated:
            print("Eser başarıyla güncellendi.")
        else:
            print("Eser bulunamadı.")

def katalogu_temizle():
    onay = input("Katalog tamamen temizlensin mi? (E/H): ")
    if onay.lower() == "e":
        with open(dosyaYolu, "w") as file:
            print("Katalog başarıyla temizlendi.")
    else:
        print("Katalog temizlenmedi.")

def kitap_ara():
    aranacak = input("Aranacak Eser Adı veya Yazar Adı: ")
    with open(dosyaYolu, "r") as file:
        kitaplar = file.readlines()
        bulunankitaplar = []
        for kitap in kitaplar:
            if aranacak.lower() in kitap.lower():
                bulunankitaplar.append(kitap.strip())
        if bulunankitaplar:
            print("Eşleşen kayıtlar:")
            for bulunankitap in bulunankitaplar:
                print(bulunankitap)
        else:
            print("Eşleşen kayıt bulunamadı.")

def main():
    while True:
        menu_goster()
        secenek = input("Bir seçenek girin (0-6): ")
        print()
        if secenek == "0":
            print("Katalog Uygulaması sonlandırılıyor...")
            break
        elif secenek == "1":
            eserleri_listele()
        elif secenek == "2":
            eser_ekle()
        elif secenek == "3":
            eser_sil()
        elif secenek == "4":
            eser_guncelle()
        elif secenek == "5":
            katalogu_temizle()
        elif secenek == "6":
            kitap_ara()
        else:
            print("Geçersiz seçenek. Tekrar deneyin.\n")

if __name__ == "__main__":
    print("Katalog Uygulaması başlatılıyor...\n")
    main()
```
